```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("/content/charity_data.csv")
application_df.head()
```

```python
application_df.drop(columns = ['EIN', 'NAME','STATUS', 'SPECIAL_CONSIDERATIONS'], inplace=True)
application_df
```

```python
app_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(app_counts[app_counts<500].index)
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
classification_counts = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = list(classification_counts[classification_counts<1000].index)
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
orgs = application_df['ORGANIZATION'].value_counts()
org_replace = list(orgs[orgs<500].index)
for org in org_replace:
    application_df['ORGANIZATION'] = application_df['ORGANIZATION'].replace(org,"Other")
application_df['ORGANIZATION'].value_counts()
```

```python
uses = application_df['USE_CASE'].value_counts()
use_replace = list(uses[uses<500].index)
for use in use_replace:
    application_df['USE_CASE'] = application_df['USE_CASE'].replace(use,"Other")
application_df['USE_CASE'].value_counts()
```

```python
affliation=application_df['AFFILIATION'].value_counts()
affliation_replace = list(affliation[affliation<500].index)
for aff in affliation_replace:
    application_df['AFFILIATION'] = application_df['AFFILIATION'].replace(aff,"Other")
application_df['AFFILIATION'].value_counts()
```

```python
income_names = ["1M-5M", "5M-10M", "10M-50M", "50M+"]
for income in income_names:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(income, "1M+")
income_names2=["1-9999", "10000-24999"]
for income in income_names2:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(income, "1-24999")

application_df['INCOME_AMT'].value_counts()
```

```python
application_df
```

```python
application_df=application_df.loc[application_df['ASK_AMT']<10000]
```

```python
dummies = pd.get_dummies(application_df)
dummies.head()
X = dummies.drop('IS_SUCCESSFUL',axis=1).values
y=dummies['IS_SUCCESSFUL'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
input_features = X_train_scaled.shape[1]
```

```python
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','sigmoid'])
    
    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value= 90,
        step=5), activation=activation, input_dim=input_features))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=30,
            step=5),
            activation=activation))
    
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return nn_model
```

```python
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=50,
    hyperband_iterations=2)
```

```python
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
top_hyper = tuner.get_best_hyperparameters(3)
for params in top_hyper:
  print(params.values)
```

```python
top_model = tuner.get_best_models(3)
for model in top_model:
    model_loss, model_accuracy = model.evaluate(X_test_scaled,y_test,verbose=2)
    print(f'Loss: {model_loss}, Accuracy: {model_accuracy}')
```
